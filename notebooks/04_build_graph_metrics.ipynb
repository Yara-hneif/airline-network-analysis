# 04 — Graph Construction & Network Metrics

## Objective
Construct a directed graph from route data and compute basic network metrics.

## Concepts Used
- Graph Theory
- Directed Graph (DiGraph)
- Degree Centrality
- Connectivity

```python
import pandas as pd
import networkx as nx
from airline_network_analysis.paths import DATA_PROCESSED_DIR
from airline_network_analysis.graph_build import build_route_digraph
```

#### 1)  Load Clean Routes

```python
routes = pd.read_csv(DATA_PROCESSED_DIR / "routes_clean.csv")
```

#### 2)  Build Directed Graph

```python
G = nx.DiGraph()

edges = list(zip(routes["src_airport_id"], routes["dst_airport_id"]))
G.add_edges_from(edges)

G.number_of_nodes(), G.number_of_edges()
```

#### 3)  Degree Centrality

    •     degree centrality top 10


```python
degree_centrality = nx.degree_centrality(G)

sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:10]
```
